```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open("log_pop_performance.csv","r")
minimum, avg, maximum = [],[],[]

for line in f:
    mini, mean, maxi = line.strip().split(",")
    minimum.append(float(mini))
    avg.append(float(mean))
    maximum.append(float(maxi))

plt.plot(range(len(maximum)),maximum, label="Maximum")
plt.plot(range(len(avg)),avg, label="Mean")
plt.plot(range(len(minimum)),minimum, label="Minimum")
plt.xlabel("ES Iteration")
plt.ylabel("Reward")
plt.legend()
plt.show()
```
